```python
%pylab inline
import oned
from draw_tree import draw_compressed
```

```python
# Set order of polynomials and threshold.
k = 5
thresh = 1.e-5

# Define some test functions
def sinx2(x):
    return sin(100. * pi * x**2) + sin(50. * pi * (-x-1)**2)

def gaussian_broad(x):
    return exp(-5. * (x-0.5)**2)

def gaussian_sharp(x):
    return exp(-50. * (x-0.5)**2)

# Create MADNESS functions
f = oned.Function(k,thresh,gaussian_broad)
g = oned.Function(k,thresh,gaussian_sharp)
h = oned.Function(k,thresh,sinx2)
```

```python
f.summarize()
g.summarize()
h.summarize()
```
